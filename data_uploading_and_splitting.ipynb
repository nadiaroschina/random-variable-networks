### Splitting by bin

We will do binning by: month, day, three hours

```python
def resample_seasonal_data(df, time_col):
    df = df.copy()
    df[time_col] = pd.to_datetime(df[time_col], format='%d.%m.%Y %H:%M', errors='coerce')
    df.set_index(time_col, inplace=True)

    numeric_cols = df.select_dtypes(include='number').columns
    categorical_cols = df.select_dtypes(include='object').columns

    def aggregate(group):
        return pd.concat([
            group[numeric_cols].mean(),
            group[categorical_cols].agg(lambda x: x.mode().iloc[0] if not x.mode().empty else None)
        ])

    df_monthly = df.resample('M').apply(aggregate)
    df_daily = df.resample('D').apply(aggregate)
    df_3h = df.resample('3H').apply(aggregate)

    return df_monthly, df_daily, df_3h
```

```python
msk_df_winter_monthly, msk_df_winter_daily, msk_df_winter_3h = resample_seasonal_data(msk_df_winter, 'Local time in Sheremetyevo (airport)')
msk_df_spring_monthly, msk_df_spring_daily, msk_df_spring_3h = resample_seasonal_data(msk_df_spring, 'Local time in Sheremetyevo (airport)')
msk_df_summer_monthly, msk_df_summer_daily, msk_df_summer_3h = resample_seasonal_data(msk_df_summer, 'Local time in Sheremetyevo (airport)')
msk_df_autumn_monthly, msk_df_autumn_daily, msk_df_autumn_3h = resample_seasonal_data(msk_df_autumn, 'Local time in Sheremetyevo (airport)')
```
